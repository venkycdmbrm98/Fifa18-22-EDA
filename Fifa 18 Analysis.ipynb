Using KNN Regressor to impute these columns

```python
def knn_impute(df, target_col):
    
    numeric_df = df.select_dtypes(numericals)
    non_na_cols = numeric_df.loc[:, numeric_df.isna().sum() == 0].columns
    
    X_train = numeric_df.loc[numeric_df[target_col].isna() == False, non_na_cols]
    y_train = numeric_df.loc[numeric_df[target_col].isna() == False, target_col]
    X_test = numeric_df.loc[numeric_df[target_col].isna() == True, non_na_cols]
    
    knn = KNeighborsRegressor()
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    
    df.loc[df[target_col].isna() == True, target_col] = y_pred
    
    return df
```

```python
cols_to_impute = ['pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic']

for col in cols_to_impute:
    knn_impute(fifa18_df, col)
```

```python
def overall_plot_num(df, cols):
    fig, axes = plt.subplots(3,4, figsize=(25,25))
    i = 0
    for row in axes:
        for ax in row:
            if i == len(cols):
                break
            sns.lineplot(data=df, y=cols[i], x='overall', ax=ax)
            i+=1
```

```python
def overall_plot_categorical(df, cols):
    fig, axes = plt.subplots(1,2, figsize=(20,8))
    plt.xticks(rotation=90)
    i = 0
    for ax in axes:
        if i == len(cols):
            break
        sns.stripplot(data=df, x=cols[i], y='overall', ax=ax, palette='magma')
        i+=1
```

### Q2. Player Roles vs Player Attributes

```python
cols_to_copy = ['player_positions', 'attacking_crossing', 'attacking_finishing', 'attacking_heading_accuracy', 'attacking_short_passing', 'attacking_volleys', 'skill_dribbling',
                'skill_curve', 'skill_fk_accuracy', 'skill_long_passing', 'skill_ball_control', 'movement_acceleration', 'movement_sprint_speed', 'movement_agility', 'movement_reactions', 
                'movement_balance', 'power_shot_power', 'power_jumping', 'power_stamina', 'power_strength', 'power_long_shots', 'mentality_aggression', 'mentality_interceptions',
                'mentality_positioning', 'mentality_vision', 'mentality_penalties', 'mentality_composure', 'defending_marking_awareness','defending_standing_tackle', 'defending_sliding_tackle',
                'goalkeeping_diving', 'goalkeeping_handling', 'goalkeeping_kicking', 'goalkeeping_positioning', 'goalkeeping_reflexes']

q2_df = fifa18_df[cols_to_copy].copy()
```

```python
q2_df['player_positions'].value_counts()
```

```python
def fn(obj):
    if (len(obj) <= 3):
        if (obj == 'GK'):
            return 'GK'
        elif (obj.endswith('B')):
            return 'DEF'
        elif (obj.endswith('M')):
            return 'MID'
        elif (obj == 'ST' or obj == 'CF' or obj.endswith('W')):
            return 'ATT'
        else:
            return 'WTF'
    else:
        ob = obj.split(',')[0]
        if (ob == 'GK'):
            return 'GK'
        elif (ob.endswith('B')):
            return 'DEF'
        elif (ob.endswith('M')):
            return 'MID'
        elif (ob == 'ST' or ob == 'CF' or ob.endswith('W')):
            return 'ATT'
        else:
            return 'WTF'
```

```python
# Create a new column for Position Area Categoryu Variable
q2_df['positional_area'] = q2_df['player_positions'].apply(fn)
```

```python
q2_df.head()
```

```python
data=q2_df.groupby('positional_area').size()
plt.figure(figsize=(6,6))
plt.pie(data, labels=data.keys(), autopct='%.0f%%')
plt.title("Percentage of players per positional area")
plt.show()
```

There are a lot of midfielders and defenders in the dataset

```python
q2_df['overall'] = fifa18_df['overall']
```

```python
mid_q2_df = q2_df[q2_df['positional_area']=='MID']
att__q2_df = q2_df[q2_df['positional_area']=='ATT']
def_q2_df = q2_df[q2_df['positional_area']=='DEF']
gk_q2_df = q2_df[q2_df['positional_area']=='GK']
```

```python
attacking_attributes = ['attacking_crossing','attacking_finishing', 'attacking_heading_accuracy', 'attacking_short_passing', 'attacking_volleys']
defending_attributes = ['defending_marking_awareness', 'defending_standing_tackle', 'defending_sliding_tackle']
goalkeeping_attributes = [ 'goalkeeping_diving', 'goalkeeping_handling', 'goalkeeping_kicking', 'goalkeeping_positioning', 'goalkeeping_reflexes']
common_attributes = ['skill_dribbling', 'skill_curve', 'skill_fk_accuracy', 'skill_long_passing', 'skill_ball_control', 'movement_acceleration', 
                     'movement_sprint_speed', 'movement_agility', 'movement_reactions', 'movement_balance', 'power_shot_power', 'power_jumping', 
                     'power_stamina', 'power_strength', 'power_long_shots', 'mentality_aggression', 'mentality_interceptions', 'mentality_positioning',
                     'mentality_vision', 'mentality_penalties', 'mentality_composure']
```

```python
def plot_positional_attributes(df, pos):
    attr = []
    attr.append(df[attacking_attributes].mean().mean())
    attr.append(df[defending_attributes].mean().mean())
    attr.append(df[goalkeeping_attributes].mean().mean())
    attr.append(df[common_attributes].mean().mean())

    labels = ['Attacking', 'Defending', 'Goalkeeping', 'Common']
    plt.figure(figsize=(7,5))
    plt.title(pos.upper(), fontdict = {'fontsize': 15})
    sns.barplot(y = attr, x = labels, palette='Blues_d', ci='sd')
    plt.xlabel('Role')
    plt.ylabel('Mean Attribute')
```

```python
plot_positional_attributes(mid_q2_df, "Midfielders")
plot_positional_attributes(att__q2_df, "Attackers")
plot_positional_attributes(def_q2_df, "Defenders")
plot_positional_attributes(gk_q2_df, "Goalkeepers")
```

As mentioned before, defenders are better at attacking than attackers are at defense

Midfielders by theory and hence proved have quite a balance in all the outfield attributes

```python
mid_22_df = q4_fifa22_df[q4_fifa22_df['positional_area']=='MID']
att_22_df = q4_fifa22_df[q4_fifa22_df['positional_area']=='ATT']
def_22_df = q4_fifa22_df[q4_fifa22_df['positional_area']=='DEF']
gk_22_df = q4_fifa22_df[q4_fifa22_df['positional_area']=='GK']


mid_18_df = q2_df[q2_df['positional_area']=='MID']
att_18_df = q2_df[q2_df['positional_area']=='ATT']
def_18_df = q2_df[q2_df['positional_area']=='DEF']
gk_18_df = q2_df[q2_df['positional_area']=='GK']
```

```python
def trait_changes(df1, df2, area):
    plt.figure(figsize=(10,7))
    trait_changes = df1.corr()['overall'] - df2.corr()['overall']
    data = trait_changes[np.abs(trait_changes) > 0.04]
    sns.lineplot(data=data, x = data.keys(), y = data.values, color = 'red')
    plt.xticks(rotation=90)
    plt.axhline(y=0, xmin=0, color='black', linestyle='dotted', linewidth=2)
    plt.title("Trait Changes from Fifa 18 to Fifa 22 for " + area)
    plt.tight_layout()
```

```python
trait_changes(att_22_df, att_18_df, "Attackers")
trait_changes(mid_22_df, mid_18_df, "Midfielders")
trait_changes(def_22_df, def_18_df, "Defenders")
trait_changes(gk_22_df, gk_18_df, "Goalkeepers")
```

Trend Change 2. Football has become more of a possession/pressing-postion-awareness based game as evident from the traits. 

Marking awareness, Vision, Stamina, Reactions, Positioning, Short and Long Passing are instrumental in deciding Overall rating which shows the change in football mentality along the years
